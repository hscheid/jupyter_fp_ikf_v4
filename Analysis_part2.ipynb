```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

We can use the numpy implementation of the poisson distrubution
\begin{equation}
f(k,\lambda) = \frac{\lambda^k}{k!}e^{-k}
\end{equation}

```python
rng = np.random.default_rng()
s = rng.poisson(1.5,100000000)
plt.hist(s,10,density = True)
plt.show()
```

```python
def fcn_poisson(k,lam):
    y = (lam**k) / np.math.factorial(k) * np.exp(-k)
    return y
```

```python
ax = range(10)
ay_lam1 = 10*[0]
ay_lam2 = 10*[0]
ay_lam5 = 10*[0]

for i in ax:
    ay_lam1[i] = fcn_poisson(i,1)
    ay_lam2[i] = fcn_poisson(i,2)
    ay_lam5[i] = fcn_poisson(i,5)

plt.plot(ax,ay_lam1,"o-")
plt.plot(ax,ay_lam2,"o-")
plt.plot(ax,ay_lam5,"o-")
plt.show()
```

```python
# we want a pdf (probability density function) so we need to normalize
# lets write a function for this
def get_pdf(y):
    s = 0
    x = y
    for i in y:
        s += i
# so we have the sum of all entries, lets divide every entry by it and we are done
    for i in range(len(x)):
        x[i] = y[i]/s
    return x
```

```python
ay_lam1 = get_pdf(ay_lam1)
ay_lam2 = get_pdf(ay_lam2)
ay_lam5 = get_pdf(ay_lam5)
```

```python
plt.plot(ax,ay_lam1,"o-")
plt.plot(ax,ay_lam2,"o-")
plt.plot(ax,ay_lam5,"o-")
plt.show()
```
